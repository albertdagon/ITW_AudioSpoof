```python
import re
import matplotlib.pyplot as plt

# Lists to store epochs and loss values
epochs = []
loss_values = []

# Open the file and parse lines
with open('AASIST_Base.txt', 'r') as file:
    for line in file:
        # Extract the epoch number
        epoch_match = re.search(r'Start training epoch(\d{3})', line)
        if epoch_match:
            # If an epoch start line is found, extract the epoch number
            current_epoch = int(epoch_match.group(1))

        # Extract the loss value after "DONE."
        loss_match = re.search(r'Loss:(\d+\.\d+)', line)
        if loss_match:
            # Append the epoch and corresponding loss to the lists
            epochs.append(current_epoch)
            loss_values.append(float(loss_match.group(1)))

# Plotting the learning curve for loss
plt.figure(figsize=(10, 6))
plt.plot(epochs, loss_values, marker='o', linestyle='-', color='b', label='Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('AASIST Learning Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

# Dictionary to store loss data for each model
models = {
    'RawGatST': 'RawGatST.txt',
    'AASIST_Base': 'AASIST_Base.txt',
    'RawNet2': 'RawNet2.txt'
}

# Initialize a dictionary to store epochs and loss values for each model
model_data = {model: {'epochs': [], 'loss': []} for model in models}

# Function to parse a file and extract epoch and loss values
def parse_file(filename):
    epochs = []
    loss_values = []
    
    with open(filename, 'r') as file:
        for line in file:
            # Extract the epoch number
            epoch_match = re.search(r'Start training epoch(\d{3})', line)
            if epoch_match:
                current_epoch = int(epoch_match.group(1))

            # Extract the loss value after "DONE."
            loss_match = re.search(r'Loss:(\d+\.\d+)', line)
            if loss_match:
                epochs.append(current_epoch)
                loss_values.append(float(loss_match.group(1)))
    
    return epochs, loss_values

# Parse each file and store the data
for model, filename in models.items():
    epochs, losses = parse_file(filename)
    model_data[model]['epochs'] = epochs
    model_data[model]['loss'] = losses

# Plotting the learning curves for each model
plt.figure(figsize=(10, 6))
for model, data in model_data.items():
    plt.plot(data['epochs'], data['loss'], marker='o', linestyle='-', label=model)

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Base Model Learning Curves')
plt.legend()
plt.grid()
plt.show()
```
